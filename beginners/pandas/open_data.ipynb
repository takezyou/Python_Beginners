# Pandasでオープンデータを扱おう

```python
# pandasの読み込み
import pandas as pd
```

### 横浜市統計ポータルサイト
http://www.city.yokohama.lg.jp/ex/stat/opendata/suikei01.html

```python
# オープンデータのURL
open_data = "http://www.city.yokohama.lg.jp/ex/stat/opendata/suikei01/e1yokohama1608.csv"
```

```python
open_data
```

```python
# データの読み込み
df = pd.read_csv(open_data, index_col='市区名')
```

```python
# データを表示
df
```

```python
# カラム名の変更
df.rename(columns={'全国地方公共団体コード':'Code'}, inplace=True)
```

```python
df
```

```python
df.tail(3)
```

```python
# 解析対象とする列の名前を指定
key_data = ['人口総数［人］', '男［人］', '女［人］']
```

```python
# 横浜市(Code: 141003)を解析対象から除外し、新しいデータフレームを作る
pp = df[df.Code!=141003][key_data]
```

```python
pp
```

```python
# 描画用モジュールの呼び出し
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
# 描画用の日本語フォントを指定
from matplotlib import rc
font = {'family':'Osaka'}
rc('font', **font)
```

```python
# 各市区に住む男女数を棒グラフで表示
pp.plot(kind='bar', figsize=(12,6), y=['男［人］', '女［人］'], fontsize=17)
```

```python
# 各市区の人口総数で並べ替え、新しいデータフレームを作成
sorted_pp = pp.sort_values(by="人口総数［人］", ascending=False)
```

```python
sorted_pp
```

```python
# 並べ替えしたデータを表示
sorted_pp.plot(kind='bar', figsize=(12,6), y=['男［人］', '女［人］'], fontsize=17)
```
